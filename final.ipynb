# The data for the votes in respect to the cities

In this part we used data preprocessing , cleneaning and formatting to achieve the desired data frame for the percantages that the top 5 most voted for parties in Turkey General Election 2018 according to their city. The original data is taken from 'YSK Açık Veri Portalı' website. The processes done on the data are the following :

The data is organized as to show the dataset of the greatest vote percantages taken in every city by the parties CHP , MHP , AKP , HDP and İYİ Parti.

````python
df_t = pd.read_excel("secim2018upd.xlsx")




# Change column names
df_t.rename(columns={'Unnamed: 0': 'Cities', 'Unnamed: 1': 'AKP', 'Unnamed: 2': 'MHP', 'Unnamed: 3': 'HDP', 'Unnamed: 4': "İYİ PARTİ", 'Unnamed: 5': "CHP"}, inplace=True)

df_t = df_t.drop(df_t.index[0])



df_t['AKP'] = df_t['AKP'].str.replace('%', '').str.split('.').str[0]
df_t['AKP'] = df_t['AKP'].fillna(0).astype(int)


df_t['MHP'] = df_t['MHP'].str.replace('%', '').str.split('.').str[0]
df_t['MHP'] = df_t['MHP'].fillna(0).astype(int)



df_t['İYİ PARTİ'] = df_t['İYİ PARTİ'].str.replace('%', '').str.split('.').str[0]
df_t['İYİ PARTİ'] = df_t['İYİ PARTİ'].fillna(0).astype(int)


df_t['HDP'] = df_t['HDP'].str.replace('%', '').str.split('.').str[0]
df_t['HDP'] = df_t['HDP'].fillna(0).astype(int)



df_t['CHP'] = df_t['CHP'].str.replace('%', '').str.split('.').str[0]
df_t['CHP'] = df_t['CHP'].fillna(0).astype(int)



df_t['AKP_ordered_city'] = df_t.apply(lambda row: f"{row['Cities']} ({row['AKP']}%)", axis=1)
df_t.sort_values(by='AKP', ascending=False, inplace=True)
akp_greatest_percantage = df_t['AKP_ordered_city']




df_t['CHP_ordered_city'] = df_t.apply(lambda row: f"{row['Cities']} ({row['CHP']}%)", axis=1)
df_t.sort_values(by='AKP', ascending=False, inplace=True)
chp_greatest_percantage = df_t['CHP_ordered_city']



df_t['MHP_ordered_city'] = df_t.apply(lambda row: f"{row['Cities']} ({row['MHP']}%)", axis=1)
df_t.sort_values(by='AKP', ascending=False, inplace=True)
mhp_greatest_percantage = df_t['MHP_ordered_city']




df_t['HDP_ordered_city'] = df_t.apply(lambda row: f"{row['Cities']} ({row['HDP']}%)", axis=1)
df_t.sort_values(by='AKP', ascending=False, inplace=True)
hdp_greatest_percantage = df_t['HDP_ordered_city']




df_t['İYİ_PARTİ_ordered_city'] = df_t.apply(lambda row: f"{row['Cities']} ({row['İYİ PARTİ']}%)", axis=1)
df_t.sort_values(by='İYİ PARTİ', ascending=False, inplace=True)
İYİPARTİ_greatest_percantage = df_t['İYİ_PARTİ_ordered_city']


greatest_percentage_df = pd.DataFrame({
    'AKP': akp_greatest_percantage,
    'CHP': chp_greatest_percantage,
    'MHP': mhp_greatest_percantage,
    'HDP': hdp_greatest_percantage,
    'İYİ PARTİ': İYİPARTİ_greatest_percantage
})




print(df_t["Cities"])
df_t=df_t.rename(columns={"Cities":"NAME_1"})
print(df_t)
sorted_df_t = df_t.sort_values("NAME_1")
sorted_df_t.head(20)

````

Visualizing relationships

````python
def plot(col1, col2, ax, df, logX=False, logY=False):
    x = df[col1]
    y = df[col2]
    if logX:
        ax.set_xscale("log")
    if logY:
        ax.set_yscale("log")

    ax.scatter(x, y)
    m, b = np.polyfit(x, y, 1)
    ax.plot(x, m*x + b, color='red')
    ax.set_title(f'{col2} vs {col1}', pad=5)
    ax.set_xlabel(col1)
    ax.set_ylabel(col2)
````

````python
fig = plt.figure(figsize=(15, 9))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)


plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plot('CHP', 'GDP per capita /National average', ax1, new_n_df)
plot('AKP', 'at least lower secondary school/ total population', ax2, new_n_df)
plot('HDP', 'GDP per capita /National average', ax3, new_n_df)
plot('HDP', 'at least upper secondary school / total population', ax4, new_n_df)

plt.show()
````

````python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Assuming that new_n_df is your DataFrame

# Select features
features = ['GDP per capita /National average', 'GDP per capita', 'at least lower secondary school/ total population', 'at least upper secondary school / total population']

# Normalize the features
scaler = MinMaxScaler()
new_n_df[features] = scaler.fit_transform(new_n_df[features])

# For each party
parties = ["AKP", "MHP", "HDP", "İYİ PARTİ", "CHP"]
models = {} # To save the models for each party
for party in parties:
    # Split the data into training and testing
    X_train, X_test, y_train, y_test = train_test_split(new_n_df[features], new_n_df[party], test_size=0.2, random_state=42)
  
    # Create a linear regression model
    model = LinearRegression()

    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f'Mean Squared Error for {party}: {mse}')

    # Save the model
    models[party] = model

# For each party
for party in parties:
    model = models[party]

    # Make predictions on the training set
    y_train_pred = model.predict(X_train)
    y_train_pred = np.round(y_train_pred).astype(int)  # Round to nearest integer

    # Calculate training accuracy
    train_accuracy = np.mean(y_train == y_train_pred)
    print(f'Training accuracy for {party}: {train_accuracy:.2f}')

    # Make predictions on the validation set
    y_test_pred = model.predict(X_test)
    y_test_pred = np.round(y_test_pred).astype(int)  # Round to nearest integer

    # Calculate validation accuracy
    validation_accuracy = np.mean(y_test == y_test_pred)
    print(f'Validation accuracy for {party}: {validation_accuracy:.2f}')    
````
